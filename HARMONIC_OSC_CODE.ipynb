```python

import numpy as np
from numpy import linalg

import matplotlib.pyplot as plt
plt.figure()
N = 500
x = np.linspace(-12,12,N)

dx = x[4]-x[3];
def kdirac(i,j):
    if(i==j):
        return 1;
    else:
        return 0;

def V0(c):
    return (1.0/4.0)*c*c;
Vpo = np.array([V0(xc) for xc in x])


def hamil(i,j):
    v = ((-kdirac(i+1,j)+2*kdirac(i,j)-kdirac(i-1,j))/((dx**2)) + kdirac(i,j)*V0(x[i]));
    return v;

Ham = np.zeros(N*N).reshape(N,N)

for i in range(N):
    for j in range(N):
        Ham[i,j] = hamil(i,j)
Eigval,Eigvect = np.linalg.eigh(Ham)



############Plotting Stuffs
plt.figure(figsize= (12,9))
plt.xlabel(r'$x$')
plt.ylabel(r'$\Psi(x)$')
N_E= 6 ## Number of Eigen vectors
S_P = 3 # scale_parameter
plt.ylim(0,Eigval[N_E])
for i in range(0,N_E):
    

    vb =  S_P*(Eigvect[:,i]**1) +Eigval[i]
    
    y = Eigval[i]*np.ones(N)
    plt.plot(x,y,'--')
    plt.plot(x,vb,label = "E(" +str(i)+") = "+str(Eigval[i]) )
plt.legend()    
plt.plot(x,Vpo)    
plt.show()    
        
```
